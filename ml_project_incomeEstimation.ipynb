```python
# Check the number of features in X_train and X_test
num_features_train = X_train.shape[1]
num_features_test = X_test.shape[1]

# Find the maximum number of features among X_train and X_test
max_num_features = max(num_features_train, num_features_test)

# If the number of features in X_train is less than max_num_features, add missing features to X_train
if num_features_train < max_num_features:
    missing_features = set(X_test.columns) - set(X_train.columns)
    for feature in missing_features:
        X_train[feature] = 0

# If the number of features in X_test is less than max_num_features, add missing features to X_test
if num_features_test < max_num_features:
    missing_features = set(X_train.columns) - set(X_test.columns)
    for feature in missing_features:
        X_test[feature] = 0
```

```python
import matplotlib.pyplot as plt

# Specify the range of maximum depths
max_depths = range(1, 21)

# Initialize lists to store the number of vertices and errors
num_vertices = []
train_errors = []
test_errors = []

# Iterate over different maximum depths
for max_depth in max_depths:
    # Create the decision tree classifier with the specified maximum depth
    tree = DecisionTreeClassifier(max_depth=max_depth)
    
    # Fit the model on the training data
    tree.fit(X_train, y_train)

    # Calculate the errors on the training, validation, and testing data
    train_error = 1 - tree.score(X_train, y_train)
    
    test_error = 1 - tree.score(X_test, y_test)

    # Append the number of vertices and errors to the respective lists
    num_vertices.append(tree.tree_.node_count)
    train_errors.append(train_error)
   
    test_errors.append(test_error)

# Plot the graph
plt.plot(num_vertices, train_errors, label='Training Error')

plt.plot(num_vertices, test_errors, label='Testing Error')
plt.xlabel('Number of Vertices')
plt.ylabel('Error')
plt.title('Number of Vertices vs. Error')
plt.legend()
plt.show()
```

```python
# Create the decision tree classifier
tree = DecisionTreeClassifier(random_state=42, max_depth=7)

# Fit the model on the training data
tree.fit(X_train, y_train)

# Prune the decision tree using reduced error pruning
path = tree.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas = path.ccp_alphas[:-1]  # Remove the maximum alpha value
trees = []
for ccp_alpha in ccp_alphas:
    pruned_tree = DecisionTreeClassifier(random_state=42, ccp_alpha=ccp_alpha, max_depth=7)
    pruned_tree.fit(X_train, y_train)
    trees.append(pruned_tree)

# Evaluate the pruned trees on the training data
train_scores = [tree.score(X_train, y_train) for tree in trees]

# Find the optimal tree based on the maximum training score
optimal_index = np.argmax(train_scores)
optimal_tree = trees[optimal_index]

# Visualize the optimal decision tree
plt.figure(figsize=(12, 6))
plot_tree(optimal_tree, filled=True, feature_names=X_train.columns)
plt.show()
```

```python
# Calculate the accuracy of the second pruned tree on the testing data
accuracy = trees[1].score(X_test, y_test)

# Print the accuracy
accuracy
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the range of values for the features
age_range = np.arange(X_train['age'].min(), X_train['age'].max()+1)
education_range = np.arange(X_train['education'].min(), X_train['education'].max()+1)
hours_per_week_range = np.arange(X_train['hours-per-week'].min(), X_train['hours-per-week'].max()+1)

# Create a meshgrid of all possible combinations of feature values
age_mesh, education_mesh, hours_per_week_mesh = np.meshgrid(age_range, education_range, hours_per_week_range)

# Flatten the meshgrid arrays
age_flattened = age_mesh.ravel()
education_flattened = education_mesh.ravel()
hours_per_week_flattened = hours_per_week_mesh.ravel()

# Create a new dataframe with the flattened feature values
X_map = pd.DataFrame({'age': age_flattened, 'workclass': 0, 'education': education_flattened, 'education-num': 0, 'marital-status': 0, 'occupation': 0, 'relationship': 0, 'race': 0, 'sex': 0, 'hours-per-week': hours_per_week_flattened, 'native-country': 0, 'age_group': 0, 'capital-gain': 0, 'capital-loss': 0})

# Select only the relevant features from X_train for training the Naive Bayes classifier
X_train_selected = X_train[['age', 'education', 'hours-per-week']]

# Fit the Naive Bayes classifier on the selected training data
naive_bayes.fit(X_train_selected, y_train)

# Predict the class probabilities for the map data using the trained Naive Bayes classifier
X_map_selected = X_map[['age', 'education', 'hours-per-week']]
y_map_prob = naive_bayes.predict_proba(X_map_selected.values)

# Reshape the predicted probabilities to match the meshgrid shape
y_map_prob_reshaped = y_map_prob[:, 1].reshape(age_mesh.shape)

# Plot the Naive Bayes map
plt.figure(figsize=(10, 8))
plt.contourf(age_mesh[:,:,0], education_mesh[:,:,0], y_map_prob_reshaped[:,:,0], cmap='coolwarm', alpha=0.8)
plt.colorbar(label='Probability of Income > $50K')
plt.xlabel('Age')
plt.ylabel('Education')
plt.title('Naive Bayes Map')
plt.show()
```
